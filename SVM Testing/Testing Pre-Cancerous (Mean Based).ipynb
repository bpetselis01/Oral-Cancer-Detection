Read Individual MATLAB Files

```julia
# Define a function to read a single MATLAB file
function read_mat_file(filename)
    mat = matread(filename)
    # do something with the contents of the MATLAB file
    return mat
end
```

Read Multiple MATLAB Files

```julia
function read_mat_files(filenames)
    data = []
    for filename in filenames
        push!(data, read_mat_file(filename))
    end
    return data
end
```

```julia
data_Q25_normal = read_mat_files(glob("Q25_*_normal*","/Users/byronpetselis/Desktop/Cancer_Data/pre"))
# Q18 & Q19 are very weird, ask Sirojan
```

```julia
feature_lesion_keys = []
feature_lesion_mean = []
feature_normal_mean = []
feature_normal_keys = []

for index in 1:15
    
    key_L = join(collect(keys(data_Q25_lesion[index])))
    value_L = collect(values(data_Q25_lesion[index]))
    mean_feature = mean(value_L[1])
    
#     println("key_L: $key_L mean: $mean_feature")
    
    push!(feature_lesion_mean, mean_feature)
    push!(feature_lesion_keys, key_L)
end


for index in 1:15
    
    key_N = join(collect(keys(data_Q25_normal[index])))
    value_N = collect(values(data_Q25_normal[index]))
    mean_feature = mean(value_N[1])
    
#     println("key_N: $key_N value_N: $value_N mean: $mean_feature")
    
    push!(feature_normal_mean, mean_feature)
    push!(feature_normal_keys, key_N)
end

# join() changes key data structure from Vector{String} to String

plot(
    bar(feature_lesion_keys, feature_lesion_mean, xlabel="Features", ylabel="Mean", legend=false), 
    xtickfontsize=5, 
    ytickfontsize=5, 
    xguidefontsize=8,
    yguidefontsize=8,
    xrotation = 45)
```

```julia
plot(
    bar(feature_normal_keys, feature_normal_mean, xlabel="Features", ylabel="Mean", legend=false), 
    xtickfontsize=5, 
    ytickfontsize=5, 
    xguidefontsize=8,
    yguidefontsize=8,
    xrotation = 45)
```

```julia
plot(feature_normal_keys, 
    [feature_normal_mean feature_lesion_mean], 
    label=["Q25 Normal" "Q25 Pre-Cancerous"], 
    xlabel="Features", 
    ylabel="Mean", 
    xrotation = 45, 
    seriestype=:scatter, 
    xtickfontsize=5, 
    ytickfontsize=5, 
    xguidefontsize=8,
    yguidefontsize=8,
    linewidth = 3)
```
